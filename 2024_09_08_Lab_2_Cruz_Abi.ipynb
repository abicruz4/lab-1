### Read in Acoustic Interferometry Data from Public GitHub Repository

Below is an example of using Pandas (pd) read_table function to read in the acoustic interferometry spectral data that was saved in Audacity and uploaded to a public GitHub repository (CHM343 > Acoustic_Interferometer > Data repository on Github).  The data is deliminated by tabs (sep="\t") and the first line is a header (header=0).

```python
air_spec = pd.read_table('https://raw.githubusercontent.com/CHM343/Acoustic_Interferometer/main/Data/2024_09_03_Air_Spectrum.txt',sep="\t", header=0)
```

### Display the data read in above using Pandas read_table function

The data was put in a variable named 'air_spec' (this can be any name you want - i.e., user defined variable).  By default, colab pandas doesn't list ALL the data when it has a large number of rows, it shows the 'head' and 'tail' of the data.

```python
air_spec
```

### Make a quick plot of the data using Pandas

Pandas is primarily a library used to do similar things to what students would use a spreadsheet (Excel) for.  You can also make basic plots.  So, lets plot the data (quickly visualize the data).



*   [Basic Pandas Plotting](https://pandas.pydata.org/docs/user_guide/visualization.html)



```python
air_spec.plot(x='Frequency (Hz)',y='Level (dB)')
```

### Visualization / Plotting using MatPlotLib Python Library

I know, I know, another python library!!  And yes, it really really is worth learning matplotlib to make high quality visualizations, plots, figures, etc in colab.  

*   [Basic Matplotlib Gallery / Examples](https://matplotlib.org/stable/gallery/index.html)


First import the matplotlib pyplot submodule component (and name it plt to make it shorter and easier to call in your code).

Then make a plot from the pandas data.


```python
import matplotlib.pyplot as plt

plt.plot(air_spec['Frequency (Hz)'],air_spec['Level (dB)'])

plt.show()
```

The plot above looks A LOT like the plot produced by Pandas.  However, matplotlib has a lot of customization and can be used to produce a high quality plot or figure or visualization.  For example, below we add a few customizations to make this plot much more descriptive (in a state that is publishable or able to add to a report).

```python
plt.plot(air_spec['Frequency (Hz)'],air_spec['Level (dB)'])
plt.xlabel('Frequency (Hz)',fontsize=16)
plt.ylabel('Level (dB)', fontsize=16)
plt.xlim(40,5000)
plt.ylim(-80,-40)
plt.show()
```

Figure Caption - This is a line plot of the acoustic interferometry spectral data from a 0.600 meter tube cavity filled with Argon gas.  This is a basic matplotlib plot and can be annotated and customized in a lot of ways to make it a better visualization.  I challenge students to be creative and try to make plots/figures that are customized to provide a clear visualization of the data and associated analysis.

### What Next?

Starting with the spectral data, you need to determine the frequency for each 'node'.  I have done this in the past using 'yet another' python library: Scipy and the 'find_peaks' function.  However, you can do this in a lot of different ways.  Often, students just estimate the frequency visually, by 'zooming in' on each node / resonance peak and visually estimating the frequency at the center of each 'node / resonance peak'.

```python
fig, (ax1, ax2) = plt.subplots(2, 1)

ax1.plot(air_spec['Frequency (Hz)'],air_spec['Level (dB)'])
ax1.set_xlim(40,5000)
ax1.set_ylim(-80,-40)
ax1.axvline(x=100,color="black", linestyle=":")
ax1.axvline(x=500,color="black", linestyle=":")

ax2.plot(air_spec['Frequency (Hz)'],air_spec['Level (dB)'])
ax2.set_xlabel('Frequency (Hz)',fontsize=12)
ax2.set_ylabel('Level (dB)', fontsize=12)
ax2.set_xlim(100,500)
ax2.set_ylim(-85,-55)
ax2.axvline(x=285,color="black", linestyle="--")

plt.show()
```

Figure Caption - This is two line plots (vertically stacked) of the acoustic interferometry spectral data from a 0.600 meter tube cavity filled with Argon gas.  The top plot shows the spectrum between 40 and 5000 Hz.  I have also included vertical dotted lines at the limits where I plot a zoomed-in plot on the bottom plot.  This zooms-in around the first node (n=1).   I then added a vertical line at an estimate of the resonance frequency (285 Hz).   The bottom plot could be zoomed into each node to estimate the resonance frequency visually.  

There is a python plotting library that is more interactive with an easier ability to zoom in and out on a plot.  This is [plotly](https://plotly.com/python/).  Yes, another python library you make want to explore!
